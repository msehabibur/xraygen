```python
# ══════════════════════════════════════════════════════════════
# OpenI Dataset: Download, Save to Drive & Visualize
# ══════════════════════════════════════════════════════════════
# Run this in Google Colab to download all OpenI chest X-ray
# images and metadata to your Google Drive.

# 1. Install dependencies
!pip install datasets pillow pandas matplotlib -q

# 2. Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# 3. Download all images + metadata
from datasets import load_dataset
from PIL import Image
import io, base64, pandas as pd
from pathlib import Path

SAVE_DIR = Path('/content/drive/MyDrive/openi_images')
SAVE_DIR.mkdir(parents=True, exist_ok=True)

print("Loading OpenI dataset...")
ds = load_dataset("ykumards/open-i", split="train")
print(f"Total rows: {len(ds)}")

saved, failed, records = 0, 0, []

for i, row in enumerate(ds):
    uid = row["uid"]
    frontal_path = lateral_path = ""

    for side in ["frontal", "lateral"]:
        try:
            img_bytes = row.get(f"img_{side}")
            if img_bytes:
                if isinstance(img_bytes, str):
                    img_bytes = base64.b64decode(img_bytes)
                img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
                p = SAVE_DIR / f"{uid}_{side}.jpg"
                img.save(p, quality=95)
                if side == "frontal": frontal_path = str(p)
                else: lateral_path = str(p)
                saved += 1
        except:
            failed += 1

    records.append({
        "uid":          uid,
        "problems":     row.get("Problems", ""),
        "findings":     row.get("findings", ""),
        "impression":   row.get("impression", ""),
        "frontal_path": frontal_path,
        "lateral_path": lateral_path,
    })

    if i % 200 == 0:
        print(f"  [{i+1}/{len(ds)}] Saved: {saved} | Failed: {failed}")

# 4. Save metadata CSV
df = pd.DataFrame(records)
df.to_csv(SAVE_DIR / "openi_metadata.csv", index=False)
print(f"\nDone! {saved} images saved to {SAVE_DIR}")

# 5. Preview sample reports
has_all = df["findings"].notna() & df["impression"].notna() & (df["frontal_path"] != "")
print(f"\nUsable for training: {has_all.sum()} samples")
print("\nSample structured report:")
row = df[has_all].iloc[0]
print(f"FINDINGS  : {row['findings']}")
print(f"IMPRESSION: {row['impression']}")

# 6. Show sample images
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
for idx, (_, r) in enumerate(df[df["frontal_path"] != ""].head(4).iterrows()):
    axes[idx].imshow(Image.open(r["frontal_path"]), cmap="gray")
    axes[idx].set_title(f"UID {r['uid']}\n{str(r['impression'])[:40]}", fontsize=8)
    axes[idx].axis("off")
plt.tight_layout()
plt.show()
```

```python
# ══════════════════════════════════════════════════════════════
# XRayGen LoRA Fine-Tuning on OpenI Dataset
# ══════════════════════════════════════════════════════════════
# Prerequisites: Run the Setup cell above to authenticate with HuggingFace.
# Hardware: Requires GPU with 40GB+ VRAM (A100 80GB recommended).

import os

# ── HPC Cache Paths (modify or remove for your environment) ──
# These paths are for Purdue Gilbreth HPC. Comment out for other systems.
os.environ["HF_HOME"] = "/scratch/gilbreth/rahma103/hf_cache"
os.environ["HF_DATASETS_CACHE"] = "/scratch/gilbreth/rahma103/hf_cache/datasets"
os.environ["TRANSFORMERS_CACHE"] = "/scratch/gilbreth/rahma103/hf_cache/models"
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

from huggingface_hub import login
HF_TOKEN = os.environ.get("HF_TOKEN", "").strip()
login(token=HF_TOKEN)

import sys, torch, base64, io, random
from torch.nn.utils.rnn import pad_sequence
from datasets import load_dataset
from transformers import AutoProcessor, AutoModelForImageTextToText, TrainingArguments, Trainer
from peft import LoraConfig, get_peft_model
from PIL import Image

# ── Config ────────────────────────────────────────────────────
MODEL_ID   = "google/medgemma-4b-it"
OUTPUT_DIR = "./medgemma-openi-lora"
HUB_MODEL  = "Habibur1003266/medgemma-roco-finetuned"
NUM_EPOCHS = 5
RESUME_CHECKPOINT = None  # Set to checkpoint path string to resume

PROMPT = (
    "You are an experienced radiologist. "
    "Analyze this chest X-ray and generate a structured report:\n"
    "FINDINGS: Describe all observable findings in clinical terms.\n"
    "IMPRESSION: Provide your overall diagnostic impression.\n"
    "RECOMMENDATION: Suggest any follow-up actions if needed.\n"
    "Respond in English only."
)

# ── Load OpenI ────────────────────────────────────────────────
print("Loading OpenI dataset...")
ds = load_dataset("ykumards/open-i", split="train")
print(f"Raw samples: {len(ds)}")

# ── Format using simple loop ─────────────────────────────────
print("Formatting dataset...")
good_samples = []

for i, row in enumerate(ds):
    try:
        findings   = (row.get("findings")   or "").strip()
        impression = (row.get("impression") or "").strip()

        if not findings and not impression:
            continue

        caption = ""
        if findings:   caption += f"FINDINGS: {findings}\n"
        if impression: caption += f"IMPRESSION: {impression}\n"
        caption += "RECOMMENDATION: Clinical correlation and follow-up as indicated."

        if not caption.isascii() or len(caption) < 20:
            continue

        image = None
        for side in ["img_frontal", "img_lateral"]:
            try:
                img_bytes = row.get(side)
                if img_bytes:
                    if isinstance(img_bytes, str):
                        img_bytes = base64.b64decode(img_bytes)
                    if len(img_bytes) > 5_000_000:
                        continue
                    img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
                    img.thumbnail((512, 512))
                    image = img
                    break
            except:
                continue

        if image is None:
            continue

        good_samples.append({
            "image":   image,
            "caption": caption.strip()
        })

    except:
        continue

    if i % 200 == 0:
        print(f"  [{i+1}/3851] Good samples: {len(good_samples)}")

print(f"\nFormatting done! {len(good_samples)} training samples")

for i in range(min(2, len(good_samples))):
    print(f"\nSample {i+1}: {good_samples[i]['caption'][:200]}")

# ── Train/Eval Split (90/10) ─────────────────────────────────
random.seed(42)
random.shuffle(good_samples)
split_idx = int(len(good_samples) * 0.9)
train_samples = good_samples[:split_idx]
eval_samples  = good_samples[split_idx:]
print(f"\nTrain: {len(train_samples)} | Eval: {len(eval_samples)}")

# ── Load Processor & Model ────────────────────────────────────
print("\nLoading processor...")
processor = AutoProcessor.from_pretrained(MODEL_ID, token=HF_TOKEN)

print("Loading model...")
model = AutoModelForImageTextToText.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16,
    device_map="auto",
    token=HF_TOKEN
)
print(f"Model on: {model.device}")

# ── Find the assistant turn boundary token IDs ────────────────
ASSISTANT_MARKER = "<start_of_turn>model\n"
assistant_marker_ids = processor.tokenizer.encode(
    ASSISTANT_MARKER, add_special_tokens=False
)
print(f"Assistant marker token IDs: {assistant_marker_ids}")

# ── LoRA ──────────────────────────────────────────────────────
peft_config = LoraConfig(
    r=32,
    lora_alpha=64,
    lora_dropout=0.05,
    target_modules=["q_proj","k_proj","v_proj","o_proj",
                    "gate_proj","up_proj","down_proj"],
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()
model.enable_input_require_grads()

# ── Helper: find subsequence in a list of token IDs ───────────
def find_subsequence(seq, subseq):
    n, m = len(seq), len(subseq)
    for i in range(n - m + 1):
        if seq[i:i+m] == subseq:
            return i
    return -1

# ── Collator ──────────────────────────────────────────────────
def collate_fn(examples):
    all_input_ids, all_attention_masks = [], []
    all_pixel_values, all_token_type_ids, all_labels = [], [], []

    for ex in examples:
        try:
            image = ex["image"]
            if image is None: continue
            if isinstance(image, Image.Image):
                image = image.convert("RGB")
            caption = ex["caption"].strip()
            if not caption: continue

            messages = [
                {"role": "user", "content": [
                    {"type": "image", "image": image},
                    {"type": "text",  "text": PROMPT}
                ]},
                {"role": "assistant", "content": [
                    {"type": "text", "text": caption}
                ]}
            ]

            prompt = processor.apply_chat_template(
                messages, tokenize=False, add_generation_prompt=False
            ).strip()

            encoded = processor(
                text=prompt, images=image,
                return_tensors="pt", padding=True
            )

            input_ids = encoded["input_ids"].squeeze(0)
            labels = input_ids.clone()

            ids_list = input_ids.tolist()
            marker_pos = find_subsequence(ids_list, assistant_marker_ids)
            if marker_pos >= 0:
                mask_end = marker_pos + len(assistant_marker_ids)
                labels[:mask_end] = -100
            labels[input_ids == processor.tokenizer.pad_token_id] = -100

            all_input_ids.append(input_ids)
            all_attention_masks.append(encoded["attention_mask"].squeeze(0))
            all_pixel_values.append(encoded["pixel_values"].squeeze(0))
            all_labels.append(labels)

            if "token_type_ids" in encoded:
                all_token_type_ids.append(encoded["token_type_ids"].squeeze(0))
            else:
                all_token_type_ids.append(torch.zeros_like(input_ids))
        except:
            continue

    if not all_input_ids:
        return None

    return {
        "input_ids":      pad_sequence(all_input_ids,      batch_first=True, padding_value=processor.tokenizer.pad_token_id),
        "attention_mask": pad_sequence(all_attention_masks, batch_first=True, padding_value=0),
        "token_type_ids": pad_sequence(all_token_type_ids,  batch_first=True, padding_value=0),
        "labels":         pad_sequence(all_labels,          batch_first=True, padding_value=-100),
        "pixel_values":   torch.stack(all_pixel_values),
    }

# ── Training Args ─────────────────────────────────────────────
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    num_train_epochs=NUM_EPOCHS,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=8,
    learning_rate=2e-4,
    warmup_ratio=0.03,
    bf16=True,
    fp16=False,
    logging_steps=25,
    save_steps=100,
    eval_strategy="steps",
    eval_steps=100,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    save_total_limit=3,
    remove_unused_columns=False,
    push_to_hub=True,
    hub_model_id=HUB_MODEL,
    report_to="none",
    dataloader_num_workers=0,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_samples,
    eval_dataset=eval_samples,
    data_collator=collate_fn,
)

# ── Train ─────────────────────────────────────────────────────
print("\nStarting training...")
print(f"   Train   : {len(train_samples)}")
print(f"   Eval    : {len(eval_samples)}")
print(f"   Epochs  : {NUM_EPOCHS}")
print(f"   Steps   : ~{len(train_samples) * NUM_EPOCHS // 16}")
print(f"   Hub     : {HUB_MODEL}\n")

trainer.train(resume_from_checkpoint=RESUME_CHECKPOINT)
trainer.save_model()
trainer.push_to_hub()
print(f"\nDone! Pushed to {HUB_MODEL}")
```

```python
# ══════════════════════════════════════════════════════════════
# XRayGen Evaluation: Base vs Fine-tuned (BERTScore + ROUGE-L)
# ══════════════════════════════════════════════════════════════
# Prerequisites: Run the Setup cell above to authenticate with HuggingFace.
# Requires: pip install rouge-score bert-score
# Hardware: Requires GPU with 40GB+ VRAM.

import os
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# ── HPC Cache Paths (modify or remove for your environment) ──
# os.environ["HF_HOME"] = "/scratch/gilbreth/rahma103/hf_cache"
# os.environ["HF_DATASETS_CACHE"] = "/scratch/gilbreth/rahma103/hf_cache/datasets"
# os.environ["TRANSFORMERS_CACHE"] = "/scratch/gilbreth/rahma103/hf_cache/models"

from huggingface_hub import login
login(token=os.environ.get("HF_TOKEN", "").strip())

import torch, random
from transformers import AutoProcessor, AutoModelForImageTextToText
from peft import PeftModel
from datasets import load_dataset
from bert_score import score as bert_score
from rouge_score import rouge_scorer
from PIL import Image
import base64, io
from datetime import datetime

# ── Config ────────────────────────────────────────────────────
MODEL_ID       = "google/medgemma-4b-it"
FINETUNED      = "Habibur1003266/medgemma-roco-finetuned"
NUM_TEST       = 100
RESULTS_FILE   = "evaluation_results_openi.txt"

# This prompt MUST match the training prompt exactly
PROMPT = (
    "You are an experienced radiologist. "
    "Analyze this chest X-ray and generate a structured report:\n"
    "FINDINGS: Describe all observable findings in clinical terms.\n"
    "IMPRESSION: Provide your overall diagnostic impression.\n"
    "RECOMMENDATION: Suggest any follow-up actions if needed.\n"
    "Respond in English only."
)

# ── Load OpenI and replicate training's exact split ───────────
print("Loading OpenI dataset...")
ds = load_dataset("ykumards/open-i", split="train")

def load_image_from_row(row):
    """Load image exactly as training does: frontal preferred, thumbnail to 512."""
    for side in ["img_frontal", "img_lateral"]:
        try:
            img_bytes = row.get(side)
            if img_bytes:
                if isinstance(img_bytes, str):
                    img_bytes = base64.b64decode(img_bytes)
                if len(img_bytes) > 5_000_000:  # Match training's 5MB filter
                    continue
                img = Image.open(io.BytesIO(img_bytes)).convert("RGB")
                img.thumbnail((512, 512))  # Match training preprocessing
                return img
        except:
            continue
    return None

def build_reference(row):
    """Build reference text matching training label format (includes RECOMMENDATION)."""
    findings   = (row.get("findings")   or "").strip()
    impression = (row.get("impression") or "").strip()
    if not findings and not impression:
        return ""
    ref = ""
    if findings:
        ref += f"FINDINGS: {findings}\n"
    if impression:
        ref += f"IMPRESSION: {impression}\n"
    # Include RECOMMENDATION to match the training label format exactly
    ref += "RECOMMENDATION: Clinical correlation and follow-up as indicated."
    return ref.strip()

# Filter samples exactly as training does
print("Filtering samples (matching training pipeline)...")
valid = []
for i, row in enumerate(ds):
    try:
        img = load_image_from_row(row)
        ref = build_reference(row)
        if img and ref and len(ref) >= 20 and ref.isascii():
            valid.append({"image": img, "reference": ref})
    except:
        continue
    if i % 500 == 0:
        print(f"  [{i+1}/3851] Valid: {len(valid)}")

print(f"Valid samples after filtering: {len(valid)}")

# Replicate training's exact shuffle + split (seed=42, 90/10)
random.seed(42)
random.shuffle(valid)
split_idx = int(len(valid) * 0.9)
train_set = valid[:split_idx]
eval_set  = valid[split_idx:]
print(f"Train split: {len(train_set)} | Eval split: {len(eval_set)}")
print(f"(Using first {NUM_TEST} from eval split for evaluation)")

test_data = eval_set[:NUM_TEST]
print(f"Test samples: {len(test_data)}")

# ── Load processor ────────────────────────────────────────────
print("\nLoading processor...")
processor = AutoProcessor.from_pretrained(MODEL_ID)

# ── Inference function ────────────────────────────────────────
def run_inference(model, image, max_tokens=400):
    messages = [{
        "role": "user",
        "content": [
            {"type": "image", "image": image},
            {"type": "text",  "text": PROMPT}
        ]
    }]
    text = processor.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )
    inputs = processor(
        text=text, images=image, return_tensors="pt"
    ).to(model.device)

    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=max_tokens,
            num_beams=4,
            early_stopping=True,
            repetition_penalty=1.2,
        )
    decoded = processor.decode(out[0], skip_special_tokens=True)
    if "model\n" in decoded:
        decoded = decoded.split("model\n")[-1].strip()
    return decoded

# ── Evaluate function ─────────────────────────────────────────
def evaluate(model, name):
    print(f"\n{'='*55}")
    print(f"Evaluating: {name}")
    print(f"{'='*55}")

    predictions, references = [], []

    for i, sample in enumerate(test_data):
        if i % 10 == 0:
            print(f"  [{i+1}/{len(test_data)}] Processing...")
        try:
            pred = run_inference(model, sample["image"])
            predictions.append(pred)
            references.append(sample["reference"])

            if i < 3:
                print(f"\n  Sample {i+1}:")
                print(f"  REF : {sample['reference'][:150]}")
                print(f"  PRED: {pred[:150]}")
        except Exception as e:
            print(f"  Sample {i} error: {e}")
            continue

    print(f"\nComputing BERTScore ({len(predictions)} samples)...")
    P, R, F1 = bert_score(
        predictions, references,
        lang="en",
        model_type="bert-base-uncased",
        verbose=False
    )

    # ROUGE-L computation
    print(f"Computing ROUGE-L ({len(predictions)} samples)...")
    rouger = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    rouge_scores = [rouger.score(ref, pred) for ref, pred in zip(references, predictions)]
    rouge_l = sum(s['rougeL'].fmeasure for s in rouge_scores) / len(rouge_scores)
    rouge_l_p = sum(s['rougeL'].precision for s in rouge_scores) / len(rouge_scores)
    rouge_l_r = sum(s['rougeL'].recall for s in rouge_scores) / len(rouge_scores)

    results = {
        "model":     name,
        "n":         len(predictions),
        "precision": P.mean().item(),
        "recall":    R.mean().item(),
        "f1":        F1.mean().item(),
        "rouge_l":   rouge_l,
        "rouge_l_p": rouge_l_p,
        "rouge_l_r": rouge_l_r,
    }
    print(f"  BERTScore P : {results['precision']:.4f}")
    print(f"  BERTScore R : {results['recall']:.4f}")
    print(f"  BERTScore F1: {results['f1']:.4f}")
    print(f"  ROUGE-L F1  : {results['rouge_l']:.4f}")
    print(f"  ROUGE-L P   : {results['rouge_l_p']:.4f}")
    print(f"  ROUGE-L R   : {results['rouge_l_r']:.4f}")
    return results, predictions

# ══════════════════════════════════════════════════════════════
# EVALUATE BASE MODEL
# ══════════════════════════════════════════════════════════════
print("\n>>> Loading BASE model...")
base_model = AutoModelForImageTextToText.from_pretrained(
    MODEL_ID, torch_dtype=torch.bfloat16, device_map="auto"
)
base_model.eval()
base_results, base_preds = evaluate(base_model, "Base MedGemma 4B")
del base_model
torch.cuda.empty_cache()

# ══════════════════════════════════════════════════════════════
# EVALUATE FINE-TUNED MODEL (OpenI)
# ══════════════════════════════════════════════════════════════
print("\n>>> Loading fine-tuned model (OpenI, 5 epochs)...")
base_for_ft = AutoModelForImageTextToText.from_pretrained(
    MODEL_ID, torch_dtype=torch.bfloat16, device_map="auto"
)
ft_model = PeftModel.from_pretrained(base_for_ft, FINETUNED)
ft_model.eval()
ft_results, ft_preds = evaluate(ft_model, "Fine-tuned (OpenI, r=32, 5 epochs)")
del ft_model
torch.cuda.empty_cache()

# ══════════════════════════════════════════════════════════════
# SUMMARY
# ══════════════════════════════════════════════════════════════
f1_delta  = ft_results["f1"]  - base_results["f1"]
f1_pct    = f1_delta / base_results["f1"] * 100
pre_delta = ft_results["precision"] - base_results["precision"]
rec_delta = ft_results["recall"]    - base_results["recall"]
rouge_delta = ft_results["rouge_l"] - base_results["rouge_l"]
rouge_pct   = rouge_delta / base_results["rouge_l"] * 100 if base_results["rouge_l"] > 0 else 0

summary = f"""
{'='*60}
XRAYGEN EVALUATION — OpenI Fine-tuned
{datetime.now().strftime('%Y-%m-%d %H:%M')}
{'='*60}

Dataset  : OpenI Indiana University Chest X-rays
Test set : {len(test_data)} samples (held-out eval split, seed=42)
Metrics  : BERTScore (bert-base-uncased), ROUGE-L
Epochs   : 5

{'─'*60}
{'Model':<40} {'P':>8} {'R':>8} {'F1':>8}
{'─'*60}
{'Base MedGemma 4B':<40} {base_results['precision']:>8.4f} {base_results['recall']:>8.4f} {base_results['f1']:>8.4f}
{'Fine-tuned (OpenI, r=32, 5ep)':<40} {ft_results['precision']:>8.4f} {ft_results['recall']:>8.4f} {ft_results['f1']:>8.4f}
{'─'*60}

ROUGE-L Scores:
{'─'*60}
{'Model':<40} {'P':>8} {'R':>8} {'F1':>8}
{'─'*60}
{'Base MedGemma 4B':<40} {base_results['rouge_l_p']:>8.4f} {base_results['rouge_l_r']:>8.4f} {base_results['rouge_l']:>8.4f}
{'Fine-tuned (OpenI, r=32, 5ep)':<40} {ft_results['rouge_l_p']:>8.4f} {ft_results['rouge_l_r']:>8.4f} {ft_results['rouge_l']:>8.4f}
{'─'*60}

IMPROVEMENT OVER BASE (BERTScore):
  F1        : {f1_delta:+.4f}  ({f1_pct:+.1f}%)
  Precision : {pre_delta:+.4f}
  Recall    : {rec_delta:+.4f}

IMPROVEMENT OVER BASE (ROUGE-L):
  F1        : {rouge_delta:+.4f}  ({rouge_pct:+.1f}%)

{'='*60}
SAMPLE PREDICTIONS (Fine-tuned, first 5):
"""

for i in range(min(5, len(ft_preds))):
    summary += f"""
Sample {i+1}:
  REFERENCE : {test_data[i]['reference'][:300]}
  PREDICTION: {ft_preds[i][:300]}
{'─'*40}"""

summary += f"\n{'='*60}\n"

print(summary)
with open(RESULTS_FILE, "w") as f:
    f.write(summary)
print(f"Results saved to {RESULTS_FILE}")
```

```python
# ══════════════════════════════════════════════════════════════
# Novel Task Demonstration: Base vs Fine-Tuned Structured Output
# ══════════════════════════════════════════════════════════════
# Proves that structured FINDINGS/IMPRESSION/RECOMMENDATION output
# is a NOVEL capability enabled by LoRA fine-tuning on OpenI.

import os, torch, random
from transformers import AutoProcessor, AutoModelForImageTextToText
from peft import PeftModel
from datasets import load_dataset
from PIL import Image
import base64, io

MODEL_ID = "google/medgemma-4b-it"
FINETUNED = "Habibur1003266/medgemma-roco-finetuned"

STRUCTURED_PROMPT = (
    "You are an experienced radiologist. "
    "Analyze this chest X-ray and generate a structured report:\n"
    "FINDINGS: Describe all observable findings in clinical terms.\n"
    "IMPRESSION: Provide your overall diagnostic impression.\n"
    "RECOMMENDATION: Suggest any follow-up actions if needed.\n"
    "Respond in English only."
)

# Load dataset and get a test sample
print("Loading OpenI dataset...")
ds = load_dataset("williamhyun/open-i", split="train")

def load_image_from_row(row):
    for col in ["image_1", "image_2"]:
        if col in row and row[col] is not None:
            if isinstance(row[col], Image.Image):
                return row[col].convert("RGB")
            if isinstance(row[col], dict) and "bytes" in row[col]:
                return Image.open(io.BytesIO(row[col]["bytes"])).convert("RGB")
    return None

# Pick 3 diverse test samples
random.seed(42)
indices = random.sample(range(len(ds)), min(200, len(ds)))
test_samples = []
for idx in indices:
    img = load_image_from_row(ds[idx])
    if img is not None:
        test_samples.append({"image": img.resize((512, 512)), "index": idx})
    if len(test_samples) >= 3:
        break

print(f"Selected {len(test_samples)} test samples")

# Load processor
processor = AutoProcessor.from_pretrained(MODEL_ID)

def run_inference(model, image, prompt, max_tokens=400):
    messages = [{"role": "user", "content": [
        {"type": "image", "image": image},
        {"type": "text", "text": prompt}
    ]}]
    text = processor.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = processor(text=text, images=image, return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(**inputs, max_new_tokens=max_tokens, num_beams=4,
                             repetition_penalty=1.2, no_repeat_ngram_size=3)
    decoded = processor.decode(out[0], skip_special_tokens=True)
    if "model\n" in decoded:
        decoded = decoded.split("model\n")[-1].strip()
    return decoded

def check_structure(text):
    """Check if output has structured sections."""
    has_findings = "FINDINGS:" in text.upper() or "FINDINGS\n" in text.upper()
    has_impression = "IMPRESSION:" in text.upper() or "IMPRESSION\n" in text.upper()
    has_recommendation = "RECOMMENDATION:" in text.upper() or "RECOMMEND" in text.upper()
    return has_findings, has_impression, has_recommendation

# ══════════════════════════════════════════════════════════════
# EVALUATE BASE MODEL
# ══════════════════════════════════════════════════════════════
print("\n" + "="*60)
print("LOADING BASE MEDGEMMA 4B (no fine-tuning)")
print("="*60)
base_model = AutoModelForImageTextToText.from_pretrained(
    MODEL_ID, torch_dtype=torch.bfloat16, device_map="auto"
)
base_model.eval()

base_outputs = []
base_structured = []
for i, sample in enumerate(test_samples):
    print(f"\n--- Base Model: Sample {i+1} ---")
    output = run_inference(base_model, sample["image"], STRUCTURED_PROMPT)
    base_outputs.append(output)
    f, im, r = check_structure(output)
    base_structured.append(all([f, im, r]))
    print(f"Output (first 300 chars):\n{output[:300]}")
    print(f"Has FINDINGS: {f} | Has IMPRESSION: {im} | Has RECOMMENDATION: {r}")
    print(f"Structured: {all([f, im, r])}")

del base_model
torch.cuda.empty_cache()

# ══════════════════════════════════════════════════════════════
# EVALUATE FINE-TUNED MODEL
# ══════════════════════════════════════════════════════════════
print("\n" + "="*60)
print("LOADING FINE-TUNED MEDGEMMA (LoRA r=32, OpenI, 5 epochs)")
print("="*60)
base_for_ft = AutoModelForImageTextToText.from_pretrained(
    MODEL_ID, torch_dtype=torch.bfloat16, device_map="auto"
)
ft_model = PeftModel.from_pretrained(base_for_ft, FINETUNED)
ft_model.eval()

ft_outputs = []
ft_structured = []
for i, sample in enumerate(test_samples):
    print(f"\n--- Fine-Tuned Model: Sample {i+1} ---")
    output = run_inference(ft_model, sample["image"], STRUCTURED_PROMPT)
    ft_outputs.append(output)
    f, im, r = check_structure(output)
    ft_structured.append(all([f, im, r]))
    print(f"Output (first 300 chars):\n{output[:300]}")
    print(f"Has FINDINGS: {f} | Has IMPRESSION: {im} | Has RECOMMENDATION: {r}")
    print(f"Structured: {all([f, im, r])}")

del ft_model
torch.cuda.empty_cache()

# ══════════════════════════════════════════════════════════════
# SUMMARY
# ══════════════════════════════════════════════════════════════
print("\n" + "="*60)
print("NOVEL TASK DEMONSTRATION: STRUCTURED OUTPUT COMPARISON")
print("="*60)
print(f"\nBase MedGemma 4B:")
print(f"  Structured output: {sum(base_structured)}/{len(base_structured)} samples")
print(f"  Success rate: {sum(base_structured)/len(base_structured)*100:.0f}%")
print(f"\nFine-Tuned MedGemma (LoRA):")
print(f"  Structured output: {sum(ft_structured)}/{len(ft_structured)} samples")
print(f"  Success rate: {sum(ft_structured)/len(ft_structured)*100:.0f}%")
print(f"\nConclusion: Structured FINDINGS/IMPRESSION/RECOMMENDATION output")
print(f"is a NOVEL capability enabled by LoRA fine-tuning on OpenI.")
print(f"The base model cannot reliably produce this clinical format.")
print("="*60)
```

```python
import os
import io
import json
import base64
import tempfile
import time
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
from PIL import Image

# ── Negation detection ────────────────────────────────────────
import re as _re

# Phrases that negate a clinical finding when they precede it
_NEGATION_PREFIXES = [
    "no ", "no evidence of ", "without ", "negative for ",
    "absence of ", "not ", "nor ", "denies ", "denied ",
    "free of ", "unremarkable for ", "no signs of ",
    "no acute ", "no significant ", "rules out ",
    "ruled out ", "no definite ", "no obvious ",
]

def _is_negated(text_lower, keyword):
    """Check if a keyword is negated in the text.
    
    Scans every occurrence of the keyword and checks whether a
    negation prefix appears in the same sentence before it.
    Returns True only if ALL occurrences are negated.
    """
    start = 0
    positions = []
    while True:
        idx = text_lower.find(keyword, start)
        if idx == -1:
            break
        positions.append(idx)
        start = idx + 1
    if not positions:
        return True  # keyword not found at all
    for idx in positions:
        # Find the start of the current sentence (look back for . ; \n)
        sent_start = max(
            text_lower.rfind('.', 0, idx),
            text_lower.rfind(';', 0, idx),
            text_lower.rfind('\n', 0, idx),
        ) + 1  # +1 to move past the delimiter (or 0 if -1+1)
        window = text_lower[sent_start:idx]
        if not any(neg in window for neg in _NEGATION_PREFIXES):
            return False  # at least one occurrence is NOT negated
    return True  # all occurrences are negated


# ── Severity keywords ─────────────────────────────────────────
SEVERITY_KW = {
    "critical": ["pneumothorax","tension pneumothorax","aortic dissection",
                 "pulmonary embolism","cardiac tamponade","hemorrhage",
                 "stroke","infarction","rupture","obstruction","hemothorax"],
    "high":     ["effusion","consolidation","mass","tumor","malignancy",
                 "cardiomegaly","pneumonia","atelectasis","fracture",
                 "edema","abscess","empyema","neoplasm"],
    "moderate": ["opacity","infiltrate","nodule","thickening","calcification",
                 "hyperinflation","scoliosis","degenerative","mild","small",
                 "minimal","borderline"],
}

SEVERITY_COLORS = {
    "critical": ("#fee2e2","#dc2626","\U0001f534"),
    "high":     ("#ffedd5","#ea580c","\U0001f7e0"),
    "moderate": ("#fef9c3","#ca8a04","\U0001f7e1"),
    "normal":   ("#dcfce7","#16a34a","\U0001f7e2"),
}

# NOTE: This prompt MUST match the training prompt used during LoRA fine-tuning
# on the OpenI dataset. Using the exact same prompt activates the learned behavior.
STRUCTURED_PROMPT = (
    "You are an experienced radiologist. "
    "Analyze this chest X-ray and generate a structured report:\n"
    "FINDINGS: Describe all observable findings in clinical terms.\n"
    "IMPRESSION: Provide your overall diagnostic impression.\n"
    "RECOMMENDATION: Suggest any follow-up actions if needed.\n"
    "Respond in English only."
)

# Chain-of-Thought prompt for enhanced reasoning (used when clinical context provided)
COT_PROMPT = (
    "You are an experienced radiologist. Think step by step:\n"
    "1. First describe what anatomical structures are visible in this chest X-ray.\n"
    "2. Then identify any abnormalities or deviations from normal.\n"
    "3. Then synthesize your FINDINGS in clinical terms.\n"
    "4. Then give your IMPRESSION with differential diagnoses if applicable.\n"
    "5. Then give your RECOMMENDATION for follow-up.\n"
    "Respond in English only."
)

# ── Confidence assessment ─────────────────────────────────────
UNCERTAIN_KEYWORDS = [
    "may", "possible", "possibly", "cannot exclude", "unclear",
    "questionable", "equivocal", "indeterminate", "cannot rule out",
    "correlate clinically", "suggest", "suspected", "probable",
]

def assess_confidence(report_text):
    """Assess report confidence based on hedging language."""
    text_lower = report_text.lower()
    found = [kw for kw in UNCERTAIN_KEYWORDS if kw in text_lower]
    if len(found) >= 3:
        return "LOW", found
    elif len(found) >= 1:
        return "MODERATE", found
    else:
        return "HIGH", []

def confidence_html(level, keywords):
    """Generate HTML badge for confidence level."""
    colors = {
        "LOW":      ("#fee2e2", "#dc2626", "\U0001f534"),
        "MODERATE": ("#fef9c3", "#ca8a04", "\U0001f7e1"),
        "HIGH":     ("#dcfce7", "#16a34a", "\U0001f7e2"),
    }
    bg, fg, dot = colors.get(level, colors["MODERATE"])
    warning = ""
    if level == "LOW":
        kw_list = ", ".join(keywords[:5])
        warning = (
            "<div style='background:#fff7ed;border:1px solid #fcd34d;border-radius:6px;"
            "padding:8px;margin-top:8px;font-size:13px;color:#92400e;'>"
            "\u26a0\ufe0f <b>Low confidence detected</b> \u2014 Radiologist review strongly recommended. "
            f"Hedging terms: {kw_list}</div>"
        )
    kw_str = f" (hedging: {", ".join(keywords[:3])})" if keywords else ""
    return (
        f"<div style='background:{bg};border-left:4px solid {fg};border-radius:8px;"
        f"padding:12px;margin-bottom:10px;font-size:14px;color:{fg};'>"  
        f"{dot} <b>Report Confidence: {level}</b>{kw_str}"
        f"{warning}</div>"
    )

def extract_icd10(report_text):
    """Extract ICD-10 codes from report using negation-aware keyword matching."""
    icd_map = {
        "pneumothorax": ("J93.9", "Pneumothorax, unspecified"),
        "pleural effusion": ("J91.8", "Pleural effusion"),
        "effusion": ("J91.8", "Pleural effusion"),
        "pneumonia": ("J18.9", "Pneumonia, unspecified"),
        "consolidation": ("J18.1", "Lobar pneumonia, unspecified"),
        "cardiomegaly": ("I51.7", "Cardiomegaly"),
        "pulmonary edema": ("J81.1", "Pulmonary edema"),
        "edema": ("J81.1", "Pulmonary edema"),
        "atelectasis": ("J98.11", "Atelectasis"),
        "fracture": ("S22.9", "Fracture of thorax, unspecified"),
        "nodule": ("R91.1", "Solitary pulmonary nodule"),
        "mass": ("R91.8", "Other nonspecific abnormal finding of lung"),
        "opacity": ("R91.8", "Other nonspecific abnormal finding of lung"),
        "calcification": ("R93.1", "Calcification of lung"),
        "scoliosis": ("M41.9", "Scoliosis, unspecified"),
        "hyperinflation": ("J43.9", "Emphysema, unspecified"),
        "copd": ("J44.1", "COPD with acute exacerbation"),
        "hernia": ("K44.9", "Diaphragmatic hernia"),
        "hiatal hernia": ("K44.9", "Diaphragmatic hernia"),
        "tuberculosis": ("A15.0", "Tuberculosis of lung"),
        "pulmonary embolism": ("I26.99", "Pulmonary embolism"),
        "aortic dissection": ("I71.0", "Dissection of aorta"),
    }
    text_lower = report_text.lower()
    found_codes = {}
    for keyword, (code, desc) in icd_map.items():
        if keyword in text_lower and code not in found_codes:
            if not _is_negated(text_lower, keyword):
                found_codes[code] = desc
    return found_codes


def icd10_html(codes):
    """Generate HTML for ICD-10 codes."""
    if not codes:
        return ""
    rows = "".join(
        f"<tr><td style='padding:6px 12px;font-family:monospace;font-weight:700;"
        f"color:#374151;border-bottom:1px solid #f3f4f6;'>{code}</td>"
        f"<td style='padding:6px 12px;color:#6b7280;border-bottom:1px solid #f3f4f6;'>{desc}</td></tr>"
        for code, desc in codes.items()
    )
    return (
        "<div style='background:#f0f9ff;border-left:4px solid #3b82f6;border-radius:8px;"
        "padding:12px;margin:10px 0;'>"
        "<b style='color:#1e40af;font-size:14px;'>\U0001f4cb Suggested ICD-10 Codes</b>"
        "<table style='width:100%;margin-top:8px;border-collapse:collapse;'>" + rows +
        "</table></div>"
    )

def assess_severity(text):
    """Negation-aware severity triage."""
    lower = text.lower()
    for level in ["critical","high","moderate"]:
        matched = [kw for kw in SEVERITY_KW[level]
                   if kw in lower and not _is_negated(lower, kw)]
        if matched:
            return level, matched
    return "normal", []


def severity_html(level, matched):
    bg, fg, emoji = SEVERITY_COLORS[level]
    kws = f" · Detected: {', '.join(matched[:4])}" if matched else ""
    return (f'<div style="display:inline-flex;align-items:center;gap:8px;'
            f'padding:10px 20px;border-radius:20px;background:{bg};'
            f'color:{fg};font-weight:700;font-size:15px;margin:8px 0;">'
            f'{emoji} Severity: {level.upper()}{kws}</div>')

def result_box(title, content, border_color="#6b7280"):
    return (f'<div style="background:white;border-radius:10px;padding:20px;'
            f'border-left:4px solid {border_color};margin:10px 0;'
            f'box-shadow:0 2px 8px rgba(0,0,0,0.06);">'
            f'<b style="color:#374151;font-size:15px;">{title}</b>'
            f'<p style="color:#4b5563;font-size:14px;line-height:1.8;'
            f'white-space:pre-wrap;margin-top:8px;">{content}</p></div>')

def translate_text(text, target_lang):
    try:
        from googletrans import Translator
        return Translator().translate(text, dest=target_lang).text
    except:
        try:
            import urllib.request, urllib.parse
            q = urllib.parse.quote(text[:500])
            url = f"https://api.mymemory.translated.net/get?q={q}&langpair=en|{target_lang}"
            with urllib.request.urlopen(url, timeout=10) as r:
                data = json.loads(r.read())
                return data["responseData"]["translatedText"]
        except:
            return "Translation unavailable."

def load_image_from_bytes(data, filename):
    if filename.lower().endswith((".dcm",".dicom")):
        try:
            import pydicom
            ds = pydicom.dcmread(io.BytesIO(data))
            arr = ds.pixel_array.astype(np.float32)
            arr = ((arr - arr.min()) / (arr.max() - arr.min() + 1e-8) * 255).astype(np.uint8)
            if arr.ndim == 2:
                arr = np.stack([arr]*3, axis=-1)
            return Image.fromarray(arr).convert("RGB")
        except Exception as e:
            return None
    return Image.open(io.BytesIO(data)).convert("RGB")

def create_download_link(content, filename, mime="text/plain"):
    if isinstance(content, str):
        content = content.encode()
    b64 = base64.b64encode(content).decode()
    uid = str(int(time.time()*1000))
    return HTML(f'''
    <a href="data:{mime};base64,{b64}" download="{filename}"
       style="display:none" id="dl_{uid}"></a>
    <script>document.getElementById("dl_{uid}").click();</script>
    ''')

# ═══════════════════════════════════════════════════════════════
# MAIN APP CLASS
# ═══════════════════════════════════════════════════════════════
class XRayGenApp:

    def __init__(self):
        self.model     = None
        self.processor = None
        self.last_report = ""
        self.history   = []
        self.current_image = None
        self._setup_widgets()

    # ── Widget setup ─────────────────────────────────────────
    def _setup_widgets(self):

        # ── PROGRESS + STATUS ────────────────────────────────
        self.progress = widgets.IntProgress(
            value=0, min=0, max=100, description="Ready:",
            bar_style="info",
            style={"bar_color":"#6b7280"},
            layout=widgets.Layout(width="500px")
        )
        self.status = widgets.HTML(
            "<div style='padding:10px;background:#f9fafb;border-left:4px solid #6b7280;"
            "border-radius:6px;font-size:14px;'><b>Status:</b> Click 'Load Model' to begin.</div>"
        )

        # ── LOAD MODEL BUTTON ─────────────────────────────────
        self.btn_load = widgets.Button(
            description="Load Model",
            layout=widgets.Layout(width="180px", height="38px"),
            style={"font_weight":"bold","button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_load.on_click(self._load_model)

        self.hf_token_input = widgets.Password(
            value=os.environ.get("HF_TOKEN", ""),
            placeholder="hf_your_token_here",
            description="HF Token:",
            style={"description_width":"80px"},
            layout=widgets.Layout(width="350px", height="38px")
        )

        self.out_load = widgets.Output()

        # ── TABS ──────────────────────────────────────────────
        self._build_tab1()
        self._build_tab2()
        self._build_tab3()
        self._build_tab4()
        self._build_tab5()

        self.tabs = widgets.Tab(
            children=[self.tab1, self.tab2, self.tab3, self.tab4, self.tab5]
        )
        for i, title in enumerate([
            "Generate Report",
            "Analyze Report",
            "Ask Questions",
            "History",
            "Agent Workflow"
        ]):
            self.tabs.set_title(i, title)

    # ── TAB BUILDERS ─────────────────────────────────────────

    def _build_tab1(self):
        """Generate report from image"""
        self.file_upload = widgets.FileUpload(
            accept=".png,.jpg,.jpeg,.dcm,.dicom",
            multiple=False,
            description="Upload Image",
            layout=widgets.Layout(width="300px")
        )
        self.file_upload.observe(self._on_file_upload, names="value")

        self.img_preview = widgets.Output(
            layout=widgets.Layout(height="280px", overflow="hidden")
        )

        self.lang_select = widgets.Dropdown(
            options=[
                ("Bengali","bn"),
                ("Hindi","hi"),
                ("Arabic","ar"),
                ("Spanish","es"),
                ("French","fr"),
            ],
            value="bn",
            description="Translate to:",
            style={"description_width":"100px"},
            layout=widgets.Layout(width="300px")
        )

        self.btn_generate = widgets.Button(
            description="Generate Report",
            disabled=True,
            layout=widgets.Layout(width="180px", height="38px"),
            style={"font_weight":"bold","button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_generate.on_click(self._generate_report)

        self.btn_download_pdf = widgets.Button(
            description="Download Report",
            disabled=True,
            layout=widgets.Layout(width="180px", height="38px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_download_pdf.on_click(self._download_report)

        self.btn_clear_upload = widgets.Button(
            description="Clear Image",
            layout=widgets.Layout(width="140px", height="38px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_clear_upload.on_click(self._clear_upload)

        self.out_generate = widgets.Output(
            layout=widgets.Layout(
                border="1px solid #e5e7eb",
                border_radius="10px",
                padding="15px",
                min_height="200px"
            )
        )

        # Clinical context input (supports MedASR voice-to-text transcription)
        self.clinical_context = widgets.Textarea(
            placeholder="Optional: Enter clinical history or paste MedASR voice transcription "
                        "(e.g., '65-year-old male, chronic cough for 3 weeks, smoker')",
            description="Clinical Context:",
            style={"description_width":"120px"},
            layout=widgets.Layout(width="440px", height="70px")
        )

        self.use_cot = widgets.Checkbox(
            value=False,
            description="Chain-of-Thought reasoning (detailed step-by-step analysis)",
            style={"description_width":"initial"},
            layout=widgets.Layout(width="440px")
        )

        left = widgets.VBox([
            widgets.HTML("<h4 style='color:#374151;font-size:16px;'>Upload Image</h4>"),
            self.file_upload,
            self.img_preview,
            self.clinical_context,
            self.use_cot,
            self.lang_select,
            widgets.HBox([self.btn_generate, self.btn_download_pdf, self.btn_clear_upload],
                         layout=widgets.Layout(gap="10px", margin="10px 0"))
        ], layout=widgets.Layout(width="480px", padding="10px"))

        right = widgets.VBox([
            widgets.HTML("<h4 style='color:#374151;font-size:16px;'>Results</h4>"),
            self.out_generate
        ], layout=widgets.Layout(padding="10px", flex="1"))

        self.tab1 = widgets.HBox(
            [left, right],
            layout=widgets.Layout(padding="15px", gap="15px")
        )

    def _build_tab2(self):
        """Analyze existing text report"""

        tab2_header = widgets.HTML("""
            <div style="background:#f9fafb;border:1px solid #e5e7eb;border-radius:10px;padding:18px;margin-bottom:15px;">
                <div style="display:flex;align-items:center;gap:10px;margin-bottom:10px;">
                    <span style="font-size:22px;">&#x1F4CB;</span>
                    <h4 style="color:#374151;font-size:16px;font-weight:700;margin:0;">Analyze an Existing Radiology Report</h4>
                </div>
                <p style="color:#6b7280;font-size:13px;margin:0;line-height:1.6;">
                    Paste a radiology report below to get a <b>severity assessment</b>, <b>plain language explanation</b>,
                    and <b>multilingual translation</b>. Click <b>Load Sample</b> to try with an example report.
                </p>
            </div>
        """)

        self.text_report_input = widgets.Textarea(
            placeholder="Paste your radiology report here...\n\nExample format:\nFINDINGS: The heart is normal in size...\nIMPRESSION: No acute cardiopulmonary disease.\nRECOMMENDATION: Clinical correlation as indicated.",
            layout=widgets.Layout(width="100%", height="200px")
        )

        self.analyze_lang = widgets.Dropdown(
            options=[
                ("Bengali","bn"),
                ("Hindi","hi"),
                ("Arabic","ar"),
                ("Spanish","es"),
                ("French","fr"),
            ],
            value="bn",
            description="Translate to:",
            style={"description_width":"100px"},
            layout=widgets.Layout(width="250px")
        )

        self.btn_analyze = widgets.Button(
            description="Analyze Report",
            disabled=True,
            layout=widgets.Layout(width="180px", height="38px"),
            style={"font_weight":"bold","button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_analyze.on_click(self._analyze_report)

        self.btn_sample_report = widgets.Button(
            description="Load Sample",
            layout=widgets.Layout(width="140px", height="38px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_sample_report.on_click(self._load_sample_report)

        self.btn_clear_report = widgets.Button(
            description="Clear",
            layout=widgets.Layout(width="100px", height="38px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_clear_report.on_click(self._clear_report)

        self.out_analyze = widgets.Output(
            layout=widgets.Layout(
                border="1px solid #e5e7eb",
                border_radius="10px",
                padding="15px",
                min_height="150px"
            )
        )

        results_header = widgets.HTML("""
            <div style="display:flex;align-items:center;gap:8px;margin-top:15px;">
                <span style="font-size:18px;">&#x1F4CA;</span>
                <h4 style="color:#374151;font-size:16px;font-weight:700;margin:0;">Results</h4>
            </div>
        """)

        self.tab2 = widgets.VBox([
            tab2_header,
            self.text_report_input,
            widgets.HBox(
                [self.analyze_lang, self.btn_analyze, self.btn_sample_report, self.btn_clear_report],
                layout=widgets.Layout(gap="10px", margin="12px 0", align_items="center")
            ),
            results_header,
            self.out_analyze
        ], layout=widgets.Layout(padding="15px"))

    def _build_tab3(self):
        """Q&A chat"""
        self.chat_history_out = widgets.Output(
            layout=widgets.Layout(
                border="1px solid #e5e7eb",
                border_radius="10px",
                padding="15px",
                min_height="250px",
                max_height="350px",
                overflow_y="auto"
            )
        )
        self.chat_input = widgets.Text(
            placeholder="Ask a question about your report...",
            layout=widgets.Layout(width="500px")
        )
        self.btn_chat = widgets.Button(
            description="Send",
            disabled=True,
            layout=widgets.Layout(width="100px", height="36px"),
            style={"font_weight":"bold","button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_chat.on_click(self._send_chat)
        self.chat_input.on_submit(self._send_chat)

        self.no_report_msg = widgets.HTML(
            "<div style='background:#fff7ed;border:1px solid #fcd34d;border-radius:10px;"
            "padding:15px;text-align:center;color:#92400e;font-size:14px;'>"
            "Please generate or analyze a report first (tabs 1 or 2).</div>"
        )

        self.tab3 = widgets.VBox([
            self.no_report_msg,
            widgets.HTML("<h4 style='color:#374151;font-size:16px;'>Chat about your report:</h4>"),
            self.chat_history_out,
            widgets.HBox(
                [self.chat_input, self.btn_chat],
                layout=widgets.Layout(gap="10px", margin="10px 0")
            )
        ], layout=widgets.Layout(padding="15px"))

        with self.chat_history_out:
            display(HTML(
                "<div style='background:#f3f4f6;border-radius:10px;padding:12px 16px;"
                "font-size:14px;margin-bottom:8px;display:inline-block;'>"
                "Hello! Generate a report first, then ask me anything about it.</div>"
            ))

    def _build_tab4(self):
        """Session history"""
        self.out_history = widgets.Output(
            layout=widgets.Layout(min_height="200px")
        )
        self.btn_clear_history = widgets.Button(
            description="Clear History",
            layout=widgets.Layout(width="160px", height="36px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_clear_history.on_click(self._clear_history)

        self.btn_export_json = widgets.Button(
            description="Export JSON",
            layout=widgets.Layout(width="160px", height="36px"),
            style={"button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_export_json.on_click(self._export_json)

        with self.out_history:
            display(HTML(
                "<p style='color:#9ca3af;text-align:center;padding:30px;font-size:15px;'>"
                "No reports generated yet.</p>"
            ))

        self.tab4 = widgets.VBox([
            widgets.HBox(
                [self.btn_clear_history, self.btn_export_json],
                layout=widgets.Layout(gap="10px", margin="0 0 15px 0")
            ),
            self.out_history
        ], layout=widgets.Layout(padding="15px"))

    # ── EVENT HANDLERS ────────────────────────────────────────

    def _load_model(self, btn):
        """Load model from HF Hub"""
        import torch
        from transformers import AutoProcessor, AutoModelForImageTextToText
        from peft import PeftModel
        from huggingface_hub import login

        token = self.hf_token_input.value.strip() or os.environ.get("HF_TOKEN", "").strip()
        if not token:
            self._set_status("Please enter your HuggingFace token first.", "#fff7ed", "#f59e0b")
            return

        self.btn_load.disabled = True
        self.progress.description = "Logging in:"
        self.progress.value = 5

        with self.out_load:
            clear_output()
            print("Logging into HuggingFace...")

        try:
            login(token=token)
            self.progress.value = 10
            self._set_status("Downloading base MedGemma 4B (~8GB)... this takes 3-5 min", "#f9fafb", "#6b7280")

            with self.out_load:
                clear_output()
                print("Loading base MedGemma 4B-IT...")

            self.processor = AutoProcessor.from_pretrained("google/medgemma-4b-it")
            self.progress.value = 40

            base_model = AutoModelForImageTextToText.from_pretrained(
                "google/medgemma-4b-it",
                torch_dtype=torch.bfloat16,
                device_map="auto",
            )
            self.progress.value = 70

            with self.out_load:
                clear_output()
                print("Loading LoRA fine-tuned adapter (OpenI)...")

            self.model = PeftModel.from_pretrained(
                base_model,
                "Habibur1003266/medgemma-roco-finetuned"
            )
            self.model.eval()
            self.progress.value = 100
            self.progress.description = "Ready:"

            device = next(self.model.parameters()).device
            with self.out_load:
                clear_output()
                print(f"Model loaded on {device}")

            self._set_status("Model ready! Upload an image to generate a report.", "#f0fdf4", "#16a34a")

            self.btn_generate.disabled = False
            self.btn_analyze.disabled  = False
            self.btn_chat.disabled     = False
            self.btn_run_agent.disabled = (self.agent_image is None)

        except Exception as e:
            with self.out_load:
                clear_output()
                print(f"Error: {e}")
            self._set_status(f"Load failed: {str(e)[:80]}", "#fee2e2", "#dc2626")
            self.btn_load.disabled = False
            self.progress.value = 0

    def _on_file_upload(self, change):
        """Preview uploaded image"""
        if not change["new"]:
            return
        for fname, fdata in change["new"].items():
            content = fdata["content"]
            img = load_image_from_bytes(content, fname)
            if img:
                self.current_image = img
                with self.img_preview:
                    clear_output(wait=True)
                    buf = io.BytesIO()
                    img.save(buf, format="PNG")
                    b64 = base64.b64encode(buf.getvalue()).decode()
                    display(HTML(
                        f'<img src="data:image/png;base64,{b64}" '
                        f'style="max-width:300px;max-height:250px;border-radius:8px;'
                        f'border:1px solid #e5e7eb;"/>'))
            break

    def _clear_upload(self, btn):
        """Clear the uploaded image"""
        self.current_image = None
        with self.img_preview:
            clear_output()

    def _clear_report(self, btn):
        """Clear the report text input and results"""
        self.text_report_input.value = ""
        with self.out_analyze:
            clear_output()

    def _load_sample_report(self, btn):
        """Load a sample radiology report for demonstration"""
        self.text_report_input.value = (
            "FINDINGS: The heart is normal in size and contour. The mediastinal "
            "and hilar contours are within normal limits. The lungs are clear "
            "without evidence of consolidation, effusion, or pneumothorax. There "
            "is no pleural abnormality. The osseous structures are intact.\n"
            "IMPRESSION: No acute cardiopulmonary disease.\n"
            "RECOMMENDATION: Clinical correlation and follow-up as indicated."
        )

    def _run_inference(self, image, prompt, max_tokens=400, min_tokens=50, use_lora=True):
        """Run MedGemma inference"""
        import torch

        # Disable LoRA for text-only tasks where the base model performs better
        if not use_lora and hasattr(self.model, 'disable_adapter_layers'):
            self.model.disable_adapter_layers()

        try:
            messages = [{
                "role": "user",
                "content": [
                    {"type": "image", "image": image},
                    {"type": "text",  "text": prompt}
                ]
            }] if image else [{
                "role": "user",
                "content": [{"type": "text", "text": prompt}]
            }]

            text = self.processor.apply_chat_template(
                messages, tokenize=False, add_generation_prompt=True
            )
            if image:
                inputs = self.processor(text=text, images=image, return_tensors="pt").to(self.model.device)
            else:
                inputs = self.processor(text=text, return_tensors="pt").to(self.model.device)

            with torch.no_grad():
                out = self.model.generate(
                    **inputs,
                    max_new_tokens=max_tokens,
                    min_new_tokens=min_tokens,
                    num_beams=4,
                    early_stopping=False,
                    repetition_penalty=1.2,
                    no_repeat_ngram_size=3,
                )
            decoded = self.processor.decode(out[0], skip_special_tokens=True)
            if "model\n" in decoded:
                decoded = decoded.split("model\n")[-1].strip()
            return decoded
        finally:
            # Re-enable LoRA adapters if they were disabled
            if not use_lora and hasattr(self.model, 'enable_adapter_layers'):
                self.model.enable_adapter_layers()

    def _generate_report(self, btn):
        """Tab 1: Generate report from image"""
        if self.current_image is None:
            with self.out_generate:
                clear_output()
                display(HTML("<p style='color:#dc2626;font-size:14px;'>Please upload an image first.</p>"))
            return

        self.btn_generate.disabled = True
        self.progress.description = "Analyzing:"
        self.progress.value = 10

        with self.out_generate:
            clear_output(wait=True)
            display(HTML(
                "<div style='text-align:center;padding:20px;'>"
                "<p style='color:#6b7280;font-size:15px;margin-top:8px;'>Analyzing image with XRayGen...</p>"
                "</div>"
            ))

        try:
            # Stage 1: Generate structured report directly from the image
            clinical_ctx = self.clinical_context.value.strip()
            if self.use_cot.value:
                prompt = COT_PROMPT
            else:
                prompt = STRUCTURED_PROMPT
            if clinical_ctx:
                prompt = f"Clinical history: {clinical_ctx}\n\n{prompt}"
            report = self._run_inference(
                self.current_image, prompt,
                max_tokens=600, min_tokens=150, use_lora=True
            )
            self.progress.value = 40
            self.last_report = report

            # Stage 2: Plain language explanation (text-only, use base model without LoRA)
            simple_prompt = (
                f"You are explaining a radiology report to a patient who has no medical background. "
                f"Based on this report, explain in simple everyday language:\n\n"
                f"- What body part was looked at and what kind of image was taken\n"
                f"- What the doctors found - is it good news, bad news, or neutral?\n"
                f"- If treatment was done (like a root canal), does it look like it went well?\n"
                f"- What the patient should do next\n"
                f"- Any reassurance if things look normal\n\n"
                f"Be warm, clear, and helpful. Use NO medical jargon. "
                f"Write at least 5-6 sentences so the patient really understands.\n\n"
                f"Report:\n{report}"
            )
            simple = self._run_inference(
                None, simple_prompt,
                max_tokens=400, min_tokens=100, use_lora=False
            )
            self.progress.value = 70

            lang = self.lang_select.value
            self.progress.description = "Translating:"
            translated = translate_text(report, lang)
            self.progress.value = 90

            sev, kws = assess_severity(report)
            conf_level, conf_kws = assess_confidence(report)
            icd_codes = extract_icd10(report)

            lang_names = {"bn":"Bengali","hi":"Hindi",
                          "ar":"Arabic","es":"Spanish","fr":"French"}

            with self.out_generate:
                clear_output(wait=True)
                display(HTML(
                    severity_html(sev, kws) +
                    confidence_html(conf_level, conf_kws) +
                    result_box("Clinical Report (English)", report, "#4b5563") +
                    icd10_html(icd_codes) +
                    result_box("Plain Language Explanation", simple, "#6b7280") +
                    result_box(f"{lang_names.get(lang,lang)} Translation", translated, "#9ca3af") +
                    "<div style='background:#fff7ed;border:1px solid #fcd34d;border-radius:8px;"
                    "padding:12px;font-size:13px;color:#92400e;margin-top:10px;'>"
                    "AI-generated for informational purposes only. Always consult a qualified physician.</div>"
                ))

            self.history.append({
                "timestamp": time.strftime("%Y-%m-%d %H:%M"),
                "severity": sev,
                "report": report,
                "simple": simple,
                "translated": translated,
            })
            self._update_history()

            self.no_report_msg.layout.display = "none"
            self.btn_download_pdf.disabled = False

            with self.chat_history_out:
                display(HTML(
                    "<div style='background:#f3f4f6;border-radius:8px;padding:10px 14px;"
                    "font-size:14px;color:#374151;margin:6px 0;'>"
                    f"Report loaded. Severity: {sev.upper()}. Ask me anything!</div>"
                ))

            self.progress.value = 100
            self.progress.description = "Done:"
            self._set_status("Report generated successfully!", "#f0fdf4", "#16a34a")

        except Exception as e:
            with self.out_generate:
                clear_output()
                display(HTML(
                    f"<div style='background:#fee2e2;border-radius:8px;padding:12px;"
                    f"color:#dc2626;font-size:14px;'>Error: {str(e)}</div>"
                ))
            self._set_status(f"Error: {str(e)[:60]}", "#fee2e2", "#dc2626")

        finally:
            self.btn_generate.disabled = False

    def _analyze_report(self, btn):
        """Tab 2: Analyze existing text report"""
        text = self.text_report_input.value.strip()
        if not text:
            with self.out_analyze:
                clear_output()
                display(HTML("<p style='color:#dc2626;font-size:14px;'>Please enter a report first.</p>"))
            return

        self.btn_analyze.disabled = True
        self.last_report = text

        with self.out_analyze:
            clear_output(wait=True)
            display(HTML(
                "<div style='text-align:center;padding:15px;'>"
                "<p style='color:#6b7280;font-size:15px;'>Analyzing...</p></div>"
            ))

        try:
            sev, kws = assess_severity(text)
            lang = self.analyze_lang.value

            simple_prompt = (
                f"You are explaining a radiology report to a patient who has no medical background. "
                f"Based on this report, explain:\n"
                f"1. What body part was examined and what type of imaging was done\n"
                f"2. What was found - is it normal or abnormal? Good news or concerning?\n"
                f"3. What does each finding mean in everyday language\n"
                f"4. What the patient should do next (if anything)\n\n"
                f"Use simple, everyday language. Avoid all medical jargon. "
                f"Be warm and clear. Write at least 4-5 sentences.\n\n"
                f"Report:\n{text}"
            )
            simple = self._run_inference(
                None, simple_prompt,
                max_tokens=400, min_tokens=100, use_lora=False
            )
            translated = translate_text(text, lang)

            lang_names = {"bn":"Bengali","hi":"Hindi",
                          "ar":"Arabic","es":"Spanish"}

            conf_level, conf_kws = assess_confidence(text)
            icd_codes = extract_icd10(text)

            with self.out_analyze:
                clear_output(wait=True)
                display(HTML(
                    severity_html(sev, kws) +
                    confidence_html(conf_level, conf_kws) +
                    icd10_html(icd_codes) +
                    result_box("Plain Language Explanation", simple, "#6b7280") +
                    result_box(f"{lang_names.get(lang,lang)} Translation", translated, "#9ca3af")
                ))

            self.no_report_msg.layout.display = "none"
            self.history.append({
                "timestamp": time.strftime("%Y-%m-%d %H:%M"),
                "severity": sev, "report": text,
                "simple": simple, "translated": translated,
            })
            self._update_history()

        except Exception as e:
            with self.out_analyze:
                clear_output()
                display(HTML(
                    f"<div style='background:#fee2e2;border-radius:8px;padding:12px;"
                    f"color:#dc2626;font-size:14px;'>Error: {str(e)}</div>"
                ))
        finally:
            self.btn_analyze.disabled = False

    def _send_chat(self, sender):
        """Tab 3: Q&A chat"""
        q = self.chat_input.value.strip()
        if not q or not self.last_report:
            return
        self.chat_input.value = ""

        with self.chat_history_out:
            display(HTML(
                f"<div style='text-align:right;margin:6px 0;'>"
                f"<span style='background:#e5e7eb;color:#374151;border-radius:12px;"
                f"padding:10px 16px;font-size:14px;display:inline-block;"
                f"border-bottom-right-radius:3px;'>{q}</span></div>"
            ))

        try:
            prompt = (
                f"You are a helpful, knowledgeable medical assistant explaining radiology results to a patient. "
                f"The patient has this radiology report:\n\n"
                f"{self.last_report}\n\n"
                f"The patient asks: {q}\n\n"
                f"Provide a detailed, helpful answer in simple language that a non-medical person can understand. "
                f"Explain what the findings mean practically for the patient. "
                f"If asked about the overall report, explain whether it looks normal or concerning, "
                f"and what the next steps might be. "
                f"Always remind them to discuss results with their doctor for personalized advice. "
                f"Give at least 3-4 sentences in your response."
            )
            answer = self._run_inference(
                None, prompt,
                max_tokens=400, min_tokens=80, use_lora=False
            )

            with self.chat_history_out:
                display(HTML(
                    f"<div style='margin:6px 0;'>"
                    f"<span style='background:#f3f4f6;color:#374151;border-radius:12px;"
                    f"padding:10px 16px;font-size:14px;display:inline-block;"
                    f"border-bottom-left-radius:3px;'>{answer}</span></div>"
                ))
        except Exception as e:
            with self.chat_history_out:
                display(HTML(f"<p style='color:#dc2626;font-size:14px;'>Error: {e}</p>"))

    def _download_report(self, btn):
        """Download text report"""
        if not self.history:
            return
        h = self.history[-1]
        content = (
            f"XRAYGEN RADIOLOGY AI REPORT\n"
            f"Generated: {h['timestamp']}\n"
            f"{'='*50}\n\n"
            f"SEVERITY: {h['severity'].upper()}\n\n"
            f"CLINICAL REPORT:\n{h['report']}\n\n"
            f"{'='*50}\n"
            f"PLAIN LANGUAGE:\n{h['simple']}\n\n"
            f"{'='*50}\n"
            f"TRANSLATION:\n{h['translated']}\n\n"
            f"{'='*50}\n"
            f"DISCLAIMER: AI-generated. Not for clinical use. Consult a physician.\n"
        )
        with self.out_generate:
            display(create_download_link(content, f"radiology_report_{int(time.time())}.txt"))

    def _update_history(self):
        with self.out_history:
            clear_output(wait=True)
            if not self.history:
                display(HTML("<p style='color:#9ca3af;text-align:center;padding:30px;font-size:15px;'>No reports yet.</p>"))
                return
            for h in reversed(self.history):
                bg, fg, emoji = SEVERITY_COLORS[h["severity"]]
                display(HTML(
                    f'<div style="background:white;border-radius:10px;padding:18px;'
                    f'margin-bottom:12px;border:0.5px solid #e5e7eb;'
                    f'box-shadow:0 2px 6px rgba(0,0,0,0.06);">'
                    f'<div style="display:flex;justify-content:space-between;align-items:center;margin-bottom:10px;">'
                    f'<span style="background:{bg};color:{fg};padding:5px 14px;border-radius:15px;font-size:13px;font-weight:700;">'
                    f'{emoji} {h["severity"].upper()}</span>'
                    f'<span style="color:#9ca3af;font-size:13px;">{h["timestamp"]}</span></div>'
                    f'<details><summary style="cursor:pointer;color:#374151;font-size:14px;font-weight:600;">View Report</summary>'
                    f'<pre style="white-space:pre-wrap;font-size:13px;color:#374151;margin-top:8px;">{h["report"][:400]}...</pre>'
                    f'<p style="font-size:13px;color:#4b5563;margin-top:6px;"><b>Plain:</b> {h["simple"][:200]}...</p>'
                    f'<p style="font-size:13px;color:#6b7280;margin-top:4px;"><b>Translation:</b> {h["translated"][:150]}...</p>'
                    f'</details></div>'
                ))

    def _clear_history(self, btn):
        self.history = []
        self._update_history()

    def _export_json(self, btn):
        content = json.dumps(self.history, indent=2)
        with self.out_history:
            display(create_download_link(content, f"reports_{int(time.time())}.json", "application/json"))

    def _set_status(self, msg, bg="#f9fafb", border="#6b7280"):
        self.status.value = (
            f"<div style='padding:10px;background:{bg};border-left:4px solid {border};"
            f"border-radius:6px;font-size:14px;'><b>Status:</b> {msg}</div>"
        )


    # ── TAB 5: AGENT WORKFLOW ────────────────────────────────

    def _build_tab5(self):
        """Agentic workflow: multi-step radiology pipeline"""

        self.agent_upload = widgets.FileUpload(
            accept=".png,.jpg,.jpeg,.dcm,.dicom",
            multiple=False,
            description="Upload Image",
            layout=widgets.Layout(width="300px")
        )
        self.agent_upload.observe(self._on_agent_upload, names="value")

        self.agent_preview = widgets.Output(
            layout=widgets.Layout(height="200px", overflow="hidden")
        )

        self.agent_context = widgets.Textarea(
            placeholder="Optional: Clinical history (e.g., '65-year-old male, chronic cough')",
            description="Context:",
            style={"description_width":"80px"},
            layout=widgets.Layout(width="440px", height="60px")
        )

        self.btn_run_agent = widgets.Button(
            description="Run Agent Workflow",
            disabled=True,
            layout=widgets.Layout(width="200px", height="38px"),
            style={"font_weight":"bold","button_color":"#f3f4f6","text_color":"#374151"}
        )
        self.btn_run_agent.on_click(self._run_agent)

        self.agent_log = widgets.Output(
            layout=widgets.Layout(
                border="1px solid #e5e7eb", border_radius="10px",
                padding="15px", min_height="300px"
            )
        )

        self.agent_image = None

        left = widgets.VBox([
            widgets.HTML(
                "<h4 style='color:#374151;font-size:16px;'>Agent Workflow</h4>"
                "<p style='color:#6b7280;font-size:13px;margin:0 0 10px 0;'>"
                "Multi-step agentic pipeline: MedGemma orchestrates 5 specialized tools "
                "with full audit trail.</p>"
            ),
            self.agent_upload,
            self.agent_preview,
            self.agent_context,
            self.btn_run_agent,
        ], layout=widgets.Layout(width="480px", padding="10px"))

        right = widgets.VBox([
            widgets.HTML("<h4 style='color:#374151;font-size:16px;'>Agent Execution Log</h4>"),
            self.agent_log
        ], layout=widgets.Layout(padding="10px", flex="1"))

        self.tab5 = widgets.HBox(
            [left, right],
            layout=widgets.Layout(padding="15px", gap="15px")
        )

    def _on_agent_upload(self, change):
        """Preview uploaded image for agent tab"""
        if not change["new"]:
            return
        try:
            name = list(change["new"].keys())[0] if isinstance(change["new"], dict) else change["new"][0]["name"]
            data = list(change["new"].values())[0]["content"] if isinstance(change["new"], dict) else change["new"][0]["content"]
            self.agent_image = load_image_from_bytes(data, name)
            with self.agent_preview:
                clear_output()
                display(self.agent_image.copy().resize((250, 250)))
            self.btn_run_agent.disabled = False
        except Exception as e:
            with self.agent_preview:
                clear_output()
                display(HTML(f"<p style='color:#dc2626;'>Error: {e}</p>"))

    def _run_agent(self, btn):
        """Execute the 5-step agentic radiology workflow"""
        if self.agent_image is None or self.model is None:
            return

        self.btn_run_agent.disabled = True
        agent_memory = []

        with self.agent_log:
            clear_output(wait=True)
            display(HTML(
                "<div style='font-family:monospace;font-size:13px;line-height:1.8;'>"
                "<b style='color:#374151;'>&#x1F916; Radiology Agent starting workflow...</b><br><br>"
                "</div>"
            ))

        def log_step(step_num, total, icon, title, detail=""):
            with self.agent_log:
                display(HTML(
                    f"<div style='font-family:monospace;font-size:13px;padding:4px 0;'>"
                    f"{icon} <b>Step {step_num}/{total}:</b> {title}"
                    f"<span style='color:#6b7280;'> {detail}</span></div>"
                ))

        try:
            clinical_ctx = self.agent_context.value.strip()
            prompt = STRUCTURED_PROMPT
            if clinical_ctx:
                prompt = f"Clinical history: {clinical_ctx}\n\n{prompt}"

            # Step 1: Analyze image (LoRA fine-tuned)
            log_step(1, 5, "&#x1F4CB;", "Analyzing chest X-ray with fine-tuned MedGemma...")
            report = self._run_inference(self.agent_image, prompt, max_tokens=600, min_tokens=150, use_lora=True)
            agent_memory.append({"step": 1, "tool": "analyze_image", "output_len": len(report)})
            self.last_report = report
            log_step(1, 5, "&#x2705;", f"Report generated", f"({len(report)} chars)")

            # Step 2: Severity triage
            log_step(2, 5, "&#x1F6A8;", "Triaging severity...")
            sev, kws = assess_severity(report)
            agent_memory.append({"step": 2, "tool": "triage_severity", "result": sev, "keywords": kws})
            log_step(2, 5, "&#x2705;", f"Severity: {sev.upper()}", f"(matched: {', '.join(kws[:3]) if kws else 'none'})")

            # Step 3: ICD-10 extraction + confidence
            log_step(3, 5, "&#x1F3E5;", "Extracting ICD-10 codes & assessing confidence...")
            icd_codes = extract_icd10(report)
            conf_level, conf_kws = assess_confidence(report)
            agent_memory.append({"step": 3, "tool": "extract_icd10_confidence", "codes": list(icd_codes.keys()), "confidence": conf_level})
            code_str = ", ".join(icd_codes.keys()) if icd_codes else "none"
            log_step(3, 5, "&#x2705;", f"ICD-10: [{code_str}]", f"| Confidence: {conf_level}")

            # Step 4: Patient simplification (base model, LoRA off)
            log_step(4, 5, "&#x1F4AC;", "Generating patient explanation (base model)...")
            simple_prompt = (
                f"You are explaining a radiology report to a patient who has no medical background. "
                f"Based on this report, explain in simple everyday language:\n"
                f"- What was found\n- Is it good news or concerning?\n- What to do next\n\n"
                f"Be warm, clear, and helpful. Use NO medical jargon. Write 4-5 sentences.\n\n"
                f"Report:\n{report}"
            )
            simplified = self._run_inference(None, simple_prompt, max_tokens=300, min_tokens=80, use_lora=False)
            agent_memory.append({"step": 4, "tool": "simplify_report", "output_len": len(simplified)})
            log_step(4, 5, "&#x2705;", f"Simplified explanation ready", f"({len(simplified)} chars)")

            # Step 5: Bengali translation
            log_step(5, 5, "&#x1F30F;", "Translating to Bengali...")
            translated = translate_text(simplified, "bn")
            agent_memory.append({"step": 5, "tool": "translate_bengali", "output_len": len(translated)})
            log_step(5, 5, "&#x2705;", "Bengali translation complete")

            # Final synthesis display
            with self.agent_log:
                display(HTML(
                    "<hr style='border:1px solid #e5e7eb;margin:15px 0;'>"
                    "<div style='font-family:monospace;font-size:13px;'>"
                    "<b>&#x1F916; AGENT WORKFLOW COMPLETE</b><br>"
                    f"Steps executed: {len(agent_memory)} | "
                    f"Severity: {sev.upper()} | "
                    f"Confidence: {conf_level} | "
                    f"ICD-10 codes: {len(icd_codes)}"
                    "</div><hr style='border:1px solid #e5e7eb;margin:15px 0;'>"
                ))
                display(HTML(
                    severity_html(sev, kws) +
                    confidence_html(conf_level, conf_kws) +
                    result_box("Clinical Report", report, "#4b5563") +
                    icd10_html(icd_codes) +
                    result_box("Patient Explanation", simplified, "#6b7280") +
                    result_box("Bengali Translation", translated, "#9ca3af")
                ))

                # Audit trail
                display(HTML(
                    "<details style='margin-top:15px;'><summary style='cursor:pointer;color:#6b7280;"
                    "font-size:13px;font-weight:600;'>&#x1F4DD; Agent Audit Trail (JSON)</summary>"
                    f"<pre style='background:#f9fafb;padding:10px;border-radius:6px;font-size:12px;'>"
                    f"{json.dumps(agent_memory, indent=2)}</pre></details>"
                ))

            self.history.append({
                "timestamp": time.strftime("%Y-%m-%d %H:%M"),
                "severity": sev, "report": report,
                "simple": simplified, "translated": translated,
                "agent_memory": agent_memory,
            })
            self._update_history()

        except Exception as e:
            with self.agent_log:
                display(HTML(
                    f"<div style='background:#fee2e2;border-radius:8px;padding:12px;"
                    f"color:#dc2626;font-size:14px;margin-top:10px;'>Agent error: {str(e)}</div>"
                ))
        finally:
            self.btn_run_agent.disabled = False

    # ── DISPLAY ───────────────────────────────────────────────
    def show(self):
        display(widgets.VBox([
            widgets.HBox(
                [self.hf_token_input, self.btn_load],
                layout=widgets.Layout(gap="10px", align_items="center", margin="15px 0 10px 0")
            ),
            widgets.HBox(
                [self.progress],
                layout=widgets.Layout(justify_content="flex-start")
            ),
            self.status,
            self.out_load,
            widgets.HTML("<hr style='border:1px solid #e5e7eb;margin:15px 0;'>"),
            self.tabs,
        ]))

# ── Launch ─────────────────────────────────────────────────────
app = XRayGenApp()
app.show()
```
